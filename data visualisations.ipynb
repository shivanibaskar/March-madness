```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#here one dataset is the regular games dataset and the other is the top games data
data1=pd.read_csv("C:\\Users\\shivani\\Desktop\\mlpresentation\\MDataFiles_Stage1\\MRegularSeasonDetailedResults.csv")
data2=pd.read_csv("C:\\Users\\shivani\\Desktop\\mlpresentation\\MDataFiles_Stage1\\MNCAATourneyDetailedResults.csv")
print(data1.shape)
print(data2.shape)
print(data1.columns)
#since both the datasets have the same number and similar columns we decide to combiner the dataset
finaldata=data1.append(data2)
datawithout2019=finaldata[finaldata['Season'] != 2019]

#we use the previous data to predict the win and loss in 2019 season so we remove 2019 from the training dataset
print(finaldata.shape)
print(datawithout2019.shape)
print(datawithout2019.head())
```

```python
import seaborn as sns
finaldata.WTeamID.value_counts()
sns.countplot(x = 'WTeamID', data = finaldata, palette = 'hls')
plt.figure(figsize=(2000, 3000))
plt.show()
```
